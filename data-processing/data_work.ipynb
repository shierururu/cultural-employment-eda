```python
df = pd.read_csv('data1.csv')
df.drop(columns=['qualifier', 'magnitude', 'indicatorId'], inplace=True)
df_foot = pd.read_csv('footnotes1.csv')
rename_dict = {
    'geoUnit': 'Страна',
    'year': 'Год',
    'value': 'Значение',
}
df.rename(columns=rename_dict, inplace=True)
df = pd.merge(df, df_foot, left_on=['Страна', 'Год'], right_on=['geoUnit', 'year'], how='left')
df.drop(columns=['geoUnit', 'year', 'indicatorId'], inplace=True)
rename_dict = {
    'type': 'Предупреждение о подсчете',
    'subtype': 'Детали предупреждения',
    'value': 'Замечание',
}
df.rename(columns=rename_dict, inplace=True)
df = df.fillna('standart')
df['Страна'] = df['Страна'].map(country_map).fillna(df['Страна'])
df = df.dropna(subset=['Значение'])
df = df[df['Значение'] > 0]
df.to_csv('dataset1.csv', index=False)
print(df.head(10))
```

```python
df = pd.read_csv('data3.csv')
df.drop(columns=['qualifier', 'magnitude'], inplace=True)
df_foot = pd.read_csv('footnotes3.csv')
rename_dict = {
    'geoUnit': 'Страна',
    'year': 'Год',
    'value': 'Значение',
}
df.rename(columns=rename_dict, inplace=True)
df = pd.merge(df, df_foot, left_on=['indicatorId', 'Страна', 'Год'], right_on=['indicatorId', 'geoUnit', 'year'], how='left')
df.drop(columns=['geoUnit', 'year'], inplace=True)
rename_dict_foot = {
    'indicatorId': 'Сфера культуры',
    'type': 'Предупреждение о подсчете',
    'subtype': 'Детали предупреждения',
    'value': 'Замечание',
}
df.rename(columns=rename_dict_foot, inplace=True)
df = df.fillna('standart')
df['Страна'] = df['Страна'].map(country_map).fillna(df['Страна'])
df['Сфера культуры'] = df['Сфера культуры'].map(code_map).fillna(df['Сфера культуры'])
df = df.sort_values(by=['Сфера культуры', 'Страна', 'Год'], ascending=[True, True, False])
df = df.dropna(subset=['Значение'])
df = df[df['Значение'] > 0]
df.to_csv('dataset3.csv', index=False)
print(df.head(10))
```
