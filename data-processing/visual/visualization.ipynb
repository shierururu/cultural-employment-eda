```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dataset1.csv')
top_countries = df.groupby('Страна')['Значение'].mean().sort_values(ascending=False).head(15)
top_countries.plot(kind='barh') 
plt.title('Топ-15 стран по значению занятых в сфере культуры')
plt.gca().invert_yaxis()
plt.savefig('chart.png')
```
